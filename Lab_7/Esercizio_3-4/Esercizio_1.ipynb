## Exercise 07.3 - 07.4
simulazioni MC NVT tramite un modello di Lennard-Jones</span> di Argon ($\sigma = 0,34$ nm, $\epsilon/k_B = 120$ K, $m=39,948$ amu) a confronto con le simulazioni AVE.

### Liquido
$\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)

```python
# Creazione di una figura con 2 subplot
fig, ax1 = plt.subplots(1, 2, figsize=(13, 4))

# Caricamento dei dati per l'energia potenziale per particella dalla simulazione MC
x_MC, actua_lMC, ave_MC, error_MC = np.loadtxt("OUTPUT/MC_sim/Liquid/potential_energy.dat", usecols=(0,1,2,3), unpack=True, skiprows=1)

# Caricamento dei dati per l'energia potenziale per particella dalla simulazione NVE
x_NVE, actual_NVE, ave_NVE, error_NVE = np.loadtxt("OUTPUT/NVE_sim/Liquid/potential_energy.dat", usecols=(0,1,2,3), unpack=True, skiprows=1)

# Grafico 1 - Energia potenziale per particella con barre di errore
ax1[0].errorbar(x_MC, ave_MC, yerr=error_MC, label="MC_sim")  # Grafico MC
ax1[0].errorbar(x_NVE, ave_NVE, yerr=error_NVE, label="NVE_sim")  # Grafico NVE
ax1[0].set_xlabel('# Blocks')  # Etichetta asse x
ax1[0].set_ylabel(r'$U/N$')  # Etichetta asse y
ax1[0].set_title(r"Potential energy per particle MC Vs AVE simulation")  # Titolo del subplot
ax1[0].grid(True)  # Attiva la griglia
ax1[0].legend()  # Mostra la legenda

# Ogni sezione del codice  commentata per chiarire cosa fa e cosa rappresenta, rendendo il codice pi comprensibile e manutenibile.


x_MC, ave_MC, error_MC = np.loadtxt("OUTPUT/MC_sim/Liquid/gofr.dat", usecols=(0,1,2),unpack=True, skiprows=1)
x_NVE, ave_NVE, error_NVE = np.loadtxt("OUTPUT/NVE_sim/Liquid/gofr.dat", usecols=(0,1,2),unpack=True, skiprows=1)

ax1[1].errorbar(x_MC,ave_MC,yerr=error_MC, label="MC_sim")
ax1[1].errorbar(x_NVE,ave_NVE,yerr=error_NVE, label="NVE_sim")
ax1[1].set_xlabel('r')
ax1[1].set_ylabel(r'$g(r)$')
ax1[1].set_title(r"g(r) MC Vs AVE simulation")
ax1[1].grid(True)
ax1[1].legend()
plt.show()

error3 = np.sqrt(error_MC**2 + error_NVE**2)
fig, ax2 = plt.subplots(1, 2, figsize=(13, 4))
ax2[0].errorbar(x_MC, ave_MC - ave_NVE, yerr=error3, label=r"$g_{MC}(r)-g_{NVE}(r)$")
ax2[0].set_xlabel('r')
ax2[0].set_ylabel(r'$g_{MC}(r)-g_{NVE}(r)$')
ax2[0].set_title(r"Difference between g(r) in MC and AVE simulation")
ax2[0].grid(True)
ax2[0].legend()

x_MC, actua_lMC, ave_MC, error_MC = np.loadtxt("OUTPUT/MC_sim/Liquid/pressure.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)
x_NVE, actual_NVE, ave_NVE, error_NVE = np.loadtxt("OUTPUT/NVE_sim/Liquid/pressure.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

ax2[1].errorbar(x_MC,ave_MC,yerr=error_MC, label="MC_sim")
ax2[1].errorbar(x_NVE,ave_NVE,yerr=error_NVE, label="NVE_sim")
ax2[1].set_xlabel('# Blocks')
ax2[1].set_ylabel(r'$P$')
ax2[1].set_title(r"Pressure MC Vs AVE simulation")
ax2[1].grid(True)
ax2[1].legend()
plt.show()
```
