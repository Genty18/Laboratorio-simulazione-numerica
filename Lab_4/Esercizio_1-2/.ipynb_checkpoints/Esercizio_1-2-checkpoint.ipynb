##### 1) Liquid argon
liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5
$)Per ottenere un sistema termalizzato alla temperatura richiesta, il valore iniziale di T è stato fissato aat **T = 1.99.**

Termalizatiees icircaut **12000 steps.**

Termalizioneoconfermataeneh"e actual temperature gra"ph.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Creating a figure with 2 subplots
fig, ax1 = plt.subplots(1, 2, figsize=(12, 4))

# Loading data from the total energy file
x, actual, ave, error = np.loadtxt("OUTPUT/Liquid/total_energy.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

# Plotting the total energy per particle with error bars
ax1[0].errorbar(x,ave,yerr=error, label="error")
ax1[0].set_xlabel('# Blocks')
ax1[0].set_ylabel(r'$E$')
ax1[0].set_title(r"Total energy per particle")
ax1[0].grid(True)
ax1[0].legend()

# Loading data from the potential energy file
x, actual, ave, error = np.loadtxt("OUTPUT/Liquid/potential_energy.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

# Plotting the potential energy per particle with error bars
ax1[1].errorbar(x,ave,yerr=error, label="error")
ax1[1].set_xlabel('# Blocks')
ax1[1].set_ylabel(r'$U$')
ax1[1].set_title(r"Potential energy  per particle")
ax1[1].grid(True)
ax1[1].legend()

# Creating a new figure with 2 subplots for kinetic energy and pressure
fig2, ax2 = plt.subplots(1, 2, figsize=(12, 4))

# Loading data from the kinetic energy file
x, actual, ave, error = np.loadtxt("OUTPUT/Liquid/kinetic_energy.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

# Plotting the kinetic energy per particle with error bars
ax2[0].errorbar(x,ave,yerr=error, label="error")
ax2[0].set_xlabel('# Blocks')
ax2[0].set_ylabel(r'$K$')
ax2[0].set_title(r"kinetic energy per particle")
ax2[0].grid(True)
ax2[0].legend()

# Loading data from the pressure file
x, actual, ave, error = np.loadtxt("OUTPUT/Liquid/pressure.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

# Plotting the pressure with error bars
ax2[1].errorbar(x,ave,yerr=error, label="error",color="red")
ax2[1].set_xlabel('# Blocks')
ax2[1].set_ylabel(r'$P$')
ax2[1].set_title(r"Pressure")
ax2[1].grid(True)
ax2[1].legend()

# Creating a new figure with 2 subplots for temperature
fig2, ax3 = plt.subplots(1, 2, figsize=(12, 4))

# Loading data from the temperature file
x, actual, ave, error = np.loadtxt("OUTPUT/Liquid/temperature.dat", usecols=(0,1,2,3),unpack=True, skiprows=1)

# Plotting the average temperature with error bars
ax3[0].errorbar(x,ave,yerr=error, label="error")
ax3[0].set_xlabel('# Blocks')
ax3[0].set_ylabel(r'$T$')
ax3[0].set_title(r"Temperature average")
ax3[0].grid(True)
ax3[0].legend()

# Plotting the actual temperature oscillations
ax3[1].plot(x, actual, label="Actual temp. oscillations", color="green")
ax3[1].set_xlabel('# Blocks')
ax3[1].set_ylabel(r'$T$')
ax3[1].set_title(r"Actual temperature")
ax3[1].grid(True)
ax3[1].legend()

# Displaying all the plots
plt.show()
```
