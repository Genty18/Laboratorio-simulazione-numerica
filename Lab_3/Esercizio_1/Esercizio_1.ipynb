### CALL option price


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 1, figsize=(10, 10))
x, f, error = np.loadtxt("DirectlyCALL.dat", usecols=(0,1,2), delimiter=' ', unpack='true')

ax[0].errorbar(x,f,yerr=error, label="error")
ax[0].set_xlabel('N')
ax[0].set_ylabel(r'$C[S(0),0]$')
ax[0].set_title(r"European CALL-option price, $C[S(0),0]$ by sampling DIRECTLY $S(T)$ as a function of the number of blocks N")
ax[0].grid(True)
ax[0].legend()

x, f, error = np.loadtxt("DiscretizedCALL.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
ax[1].errorbar(x,f,yerr=error, label="error")
ax[1].set_xlabel('N')
ax[1].set_ylabel(r'$C[S(0),0]$')
ax[1].set_title(r"European CALL-option price, $C[S(0),0]$ By sampling the discretized $GMB(r, \sigma^2)$ as a function of the number of blocks N")
ax[1].grid(True)
ax[1].legend()

plt.show()
```
