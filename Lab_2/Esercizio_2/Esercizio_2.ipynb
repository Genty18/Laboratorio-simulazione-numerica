### 3D Random Walks (RW) on a cubic lattice

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#definizione gaussiana per il fit
def f(x,a):  
    return a*np.sqrt(x)

fig, ax = plt.subplots(1, 1, figsize=(10, 7))
x, values, error = np.loadtxt("dataDISC.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,values,yerr=error*10, label='Data')

# Esegui il fit dei dati
popt, pcov = curve_fit(f, x, values)

# Visualizza i dati originali e il fit
plt.plot(x, f(x, *popt), 'r-', label='Fit: a=%5.3f' % tuple(popt))

plt.xlabel('#Step')
plt.ylabel(r'Lenght')
plt.title(r"Lattice RW's lenght as function of step number")
plt.grid(True)
plt.legend()

plt.show()
```

#### *Ivalori degli errori sono stati moltiplicati per 10, per visualizzarli meglio 

## 3D Random Walks (RW) in the continuum:
È importante notare che per poter campionare uniformemente un punto su una sfera di raggio r, l'angolo $\phi$ deve essere campionato uniformemente in [0, $2\pi$], mentre per campionare l'angolo $\theta$, va campionato uniformemente il $cos(\theta)$ tra [-1,1], perchè va considerato lo jacobiano delle coordinate sferiche

```python
#definizione gaussiana per il fit
def f(x,a):  
    return a*np.sqrt(x)

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

x, values, error = np.loadtxt("dataCONT.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,values,yerr=error*10, label='Data')
# Esegui il fit dei dati
popt, pcov = curve_fit(f, x, values)
# Visualizza i dati originali e il fit
plt.plot(x, f(x, *popt), 'r-', label='Fit: a=%5.3f' % tuple(popt))

plt.xlabel('#Step')
plt.ylabel(r'Lenght')
plt.title(r"Continuum RW's lenght as function of step number")
plt.grid(True)
plt.legend()

plt.show()
```
