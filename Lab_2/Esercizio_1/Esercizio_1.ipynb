### Exercise 02.1
We wont to compute the following 1D integral via Uniform and Importance sampling of Monte Carlo methods.
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

## Uniform samling


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("Uniform_sampling.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#throws')
plt.ylabel(r'I')
plt.title(r"Values of I as function of Throws")
plt.grid(True)
plt.show()
```
