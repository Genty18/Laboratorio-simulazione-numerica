**1D Ising model**.

Hamiltonian of the system : $$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

### I) Energia interna (con $h=0$):
plot di $U(N,T) = \langle H \rangle_{h=0}$ in funzione di $T$ nell'intervallo $\left[0.5;2.0\right]$ per un sistema di $N=50$ spin con $J=1$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Creazione di una figura con un solo subplot
fig, ax1 = plt.subplots(1, 1, figsize=(12, 8))

# Intervallo di temperatura e passo
t1 = 3.0
t2 = 0.5
deltaT = 0.0625
step = int((t1 - t2) / deltaT)  # Numero di passi tra t2 e t1

# Creazione di un array di temperature generate linearmente tra t2 e t1
T = np.linspace(t2, t1, step)

# Inizializzazione delle liste per memorizzare i dati
M = []      # Energia totale media per Metropolis
G = []      # Energia totale media per Gibbs
err_M = []  # Errori associati a Metropolis
err_G = []  # Errori associati a Gibbs

# Caricamento dei dati dell'energia totale dai file simulati con Metropolis
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Metropolis/H=0/SIM{i}/total_energy.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    M.append(ave)           # Aggiunta dell'energia totale media alla lista M
    err_M.append(error * 50)  # Aggiunta dell'errore scalato alla lista err_M

# Caricamento dei dati dell'energia totale dai file simulati con Gibbs
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Gibbs/H=0/SIM{i}/total_energy.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    G.append(ave)           # Aggiunta dell'energia totale media alla lista G
    err_G.append(error * 50)  # Aggiunta dell'errore scalato alla lista err_G

# Calcolo dei valori teorici dell'energia totale per particella
beta = 1 / T
J = 1.0
Ns = 50
th = np.tanh(J / T)
thN = th ** Ns
ch = 1 / th
e = -J * (th + ch * thN) / (1 + thN)

# Aggiunta della curva teorica al grafico
ax1.plot(T, e, label="Theoretical")

# Tracciamento dell'energia totale per particella con barre di errore per i dati di Metropolis
ax1.errorbar(T, M, yerr=err_M, fmt='x', label="Metropolis sim", color="red")

# Tracciamento dell'energia totale per particella con barre di errore per i dati di Gibbs
ax1.errorbar(T, G, yerr=err_G, fmt='+', label="Gibbs sim", color="orange")

# Configurazione degli assi e delle etichette del grafico
ax1.set_xlabel('T')
ax1.set_ylabel(r'$E_{tot}$')
ax1.set_title(r"1D Ising model : Total energy")
ax1.grid(True)  # Aggiunta di una griglia al grafico
ax1.legend()    # Aggiunta di una legenda al grafico

# Mostra il grafico
plt.show()
```

*Il valore delle barre di errore delle simulazioni metropoli e Gibbs **è stato moltiplicato per 50** per poterlo visualizzare correttamente sul grafico

### II) Capacità termica (con $h=0$):plot die $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ in funzione di $T$ in l'intervallo $\left[0.5;2.0\right]$ per un sistema di $N=50$ spina con $J=1$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Creazione di una figura con un solo subplot
fig, ax1 = plt.subplots(1, 1, figsize=(12, 7))

# Definizione dei parametri di temperatura
t1 = 3.0
t2 = 0.5
deltaT = 0.0625
step = int((t1 - t2) / deltaT)  # Numero di passi tra t2 e t1

# Creazione di un array di temperature generate linearmente tra t2 e t1
T = np.linspace(t2, t1, step)

# Inizializzazione delle liste per memorizzare i dati
M = []      # Specific heat per Metropolis
G = []      # Specific heat per Gibbs
err_M = []  # Errori associati a Metropolis
err_G = []  # Errori associati a Gibbs

# Caricamento dei dati del calore specifico da file per Metropolis
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Metropolis/H=0/SIM{i}/specific_heat.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    M.append(ave)           # Aggiunta del calore specifico medio alla lista M
    err_M.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_M

# Caricamento dei dati del calore specifico da file per Gibbs
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Gibbs/H=0/SIM{i}/specific_heat.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    G.append(ave)           # Aggiunta del calore specifico medio alla lista G
    err_G.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_G

# Calcolo dei valori teorici del calore specifico
beta = 1 / T
J = 1.0
Ns = 50
th = np.tanh(J / T)
thN = th ** Ns
ch = 1 / th
heat = ((beta * J) ** 2) * (((1 + thN + (Ns - 1) * (th ** 2) + (Ns - 1) * (ch ** 2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN)) ** 2)

# Tracciamento della curva teorica del calore specifico
ax1.plot(T, heat * 50, label="Theoretical")

# Tracciamento del calore specifico con barre di errore per i dati di Metropolis
ax1.errorbar(T, M, yerr=err_M, fmt='x', label="Metropolis sim", color="red")

# Tracciamento del calore specifico con barre di errore per i dati di Gibbs
ax1.errorbar(T, G, yerr=err_G, fmt='+', label="Gibbs sim", color="orange")

# Configurazione degli assi e delle etichette del grafico
ax1.set_xlabel('T')
ax1.set_ylabel(r'$C_{v}$')
ax1.set_title(r"1D Ising model : Heat capacity")
ax1.grid(True)  # Aggiunta di una griglia al grafico
ax1.legend()    # Aggiunta di una legenda al grafico

# Mostra il grafico
plt.show()
```

*Il valore delle barre di errore delle simulazioni metropoli e Gibbs **è stato moltiplicato per 15** per poterlo visualizzare correttamente sul grafico

### III) Magnetizzazione (con $h \not= 0$):
Plot $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$ in funzione di $T$ nell'intervallo $\left[0.5;2.0 \right]$ per un sistema di $N=50$ spin con $J=1$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Creazione di una figura con un solo subplot
fig, ax1 = plt.subplots(1, 1, figsize=(12, 7))

# Definizione dei parametri di temperatura
t1 = 3.0
t2 = 0.5
deltaT = 0.0625
step = int((t1 - t2) / deltaT)  # Numero di passi tra t2 e t1

# Creazione di un array di temperature generate linearmente tra t2 e t1
T = np.linspace(t2, t1, step)

# Inizializzazione delle liste per memorizzare i dati
M = []      # Suscettibilità magnetica per Metropolis
G = []      # Suscettibilità magnetica per Gibbs
err_M = []  # Errori associati a Metropolis
err_G = []  # Errori associati a Gibbs

# Caricamento dei dati della suscettibilità magnetica da file per Metropolis
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Metropolis/H=0/SIM{i}/susceptibility.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    M.append(ave)           # Aggiunta della suscettibilità magnetica media alla lista M
    err_M.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_M

# Caricamento dei dati della suscettibilità magnetica da file per Gibbs
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Gibbs/H=0/SIM{i}/susceptibility.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    G.append(ave)           # Aggiunta della suscettibilità magnetica media alla lista G
    err_G.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_G

# Calcolo dei valori teorici della suscettibilità magnetica
beta = 1 / T
J = 1.0
Ns = 50
th = np.tanh(J / T)
thN = th ** Ns
X = beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)

# Tracciamento della curva teorica della suscettibilità magnetica
ax1.plot(T, X * 50, label="Theoretical")

# Tracciamento della suscettibilità magnetica con barre di errore per i dati di Metropolis
ax1.errorbar(T, M, yerr=err_M, fmt='x', label="Metropolis sim", color="red")

# Tracciamento della suscettibilità magnetica con barre di errore per i dati di Gibbs
ax1.errorbar(T, G, yerr=err_G, fmt='+', label="Gibbs sim", color="orange")

# Configurazione degli assi e delle etichette del grafico
ax1.set_xlabel('T')
ax1.set_ylabel(r'$M$')
ax1.set_title(r"1D Ising model : Magnetization")
ax1.grid(True)  # Aggiunta di una griglia al grafico
ax1.legend()    # Aggiunta di una legenda al grafico

# Mostra il grafico
plt.show()
```

*Il valore delle barre di errore delle simulazioni metropoli e Gibbs **è stato moltiplicato per 15** per poterlo visualizzare correttamente sul grafico

### IV) Suscettività magnetica (con $h=0$):Plot di o $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$ in funzione di $T$ nell'intervallo $\left[ 0.5;2.0\right]$ per un sistema dii $N=50$ spin  con $J=1$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Creazione di una figura con un solo subplot
fig, ax1 = plt.subplots(1, 1, figsize=(12, 7))

# Definizione dei parametri di temperatura
t1 = 3.0
t2 = 0.5
deltaT = 0.0625
step = int((t1 - t2) / deltaT)  # Numero di passi tra t2 e t1

# Creazione di un array di temperature generate linearmente tra t2 e t1
T = np.linspace(t2, t1, step)

# Inizializzazione delle liste per memorizzare i dati
M = []      # Magnetizzazione per Metropolis
G = []      # Magnetizzazione per Gibbs
err_M = []  # Errori associati a Metropolis
err_G = []  # Errori associati a Gibbs

# Caricamento dei dati della magnetizzazione da file per Metropolis
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Metropolis/H=0.02/SIM{i}/magnetization.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    M.append(ave)           # Aggiunta della magnetizzazione media alla lista M
    err_M.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_M

# Caricamento dei dati della magnetizzazione da file per Gibbs
for i in range(step):
    x, actual, ave, error = np.loadtxt(f"OUTPUT/Gibbs/H=0.02/SIM{i}/magnetization.dat", usecols=(0, 1, 2, 3), unpack=True, skiprows=20)
    G.append(ave)           # Aggiunta della magnetizzazione media alla lista G
    err_G.append(error * 15)  # Aggiunta dell'errore scalato alla lista err_G

# Calcolo dei valori teorici della suscettibilità magnetica
h = 0.02  # Campo esterno
b = 1 / T
J = 1.0
Ns = 50
l1 = np.exp(b * J) * np.cosh(b * h) + np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
l2 = np.exp(b * J) * np.cosh(b * h) - np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
Z = l1 ** Ns + l2 ** Ns
Ma = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

# Tracciamento della curva teorica della suscettibilità magnetica
ax1.plot(T, Ma * 50, label="Theoretical")

# Tracciamento della magnetizzazione con barre di errore per i dati di Metropolis
ax1.errorbar(T, M, yerr=err_M, fmt='x', label="Metropolis sim", color="red")

# Tracciamento della magnetizzazione con barre di errore per i dati di Gibbs
ax1.errorbar(T, G, yerr=err_G, fmt='+', label="Gibbs sim", color="orange")

# Configurazione degli assi e delle etichette del grafico
ax1.set_xlabel('T')
ax1.set_ylabel(r'$\chi$')  # Etichetta per la suscettibilità magnetica
ax1.set_title(r"1D Ising model : Magnetic susceptibility ")  # Titolo del grafico
ax1.grid(True)  # Aggiunta di una griglia al grafico
ax1.legend()    # Aggiunta di una legenda al grafico

# Mostra il grafico
plt.show()
```
