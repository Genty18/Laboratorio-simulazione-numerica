### Stima di $<r>$ e della sua incertezza come funzione del numero di lanci
r is a uniformly distribuited random variable, in order to test the Pseudo-Random Number generator.


We expect a mean value of 0.5 : $\langle r \rangle = \int_0^1 r dr = 1/2$.

The statistical uncertainties has been evaluated with the blocking method, seen in the lectures. The "Media.dat" file, containing the data has been produced by the code in "main.cpp" file. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("Media.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, label='data')
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.title("<r> - 1/2 wiht errors")
plt.legend()
plt.grid(True)
plt.show()
```

### Chi quadro
In order to estimate the probability that r  are not drawn from a uniform distribution, the Pearson's cumulative test statistic $\chi^2$ has been applied. 100 values of $\chi^2$ are plotted below.

In our case the interval [0,1] has been divided in 100(=M) sub intervals and has been count how many of 10000(=N) throws have been fallen in each sub intervals. 

chi-square statistic int this case:
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - N/M \right)^2}{N/M}$$

Expected events $Np = N\times 1/M= N/M$ with $(n_i - n/M)^2 \simeq n/M$ and thus $\chi^2 \simeq 100$.

The "Chi2.dat" file, containing the data has been produced by the code in "Chi2.cpp" file.

n

```python
from scipy.stats import chi2

omega = np.loadtxt("Chi2.dat", usecols=(0), delimiter=' ', unpack='true')
bins = np.linspace(40, 160, 50)
plt.hist(
    omega,
    bins=bins,
    density=True
);
y = chi2.pdf(bins, 100)
plt.plot(
    bins, y,
    color="red",
    lw=2,
    label="Theoretical"
)
plt.xlabel(r'$\chi^2$')
plt.ylabel(r'$\rho(\chi^2)$')
plt.title(r'Distribution of 100 values of $\chi^2$')
plt.legend(loc=(0.110, 0.7), frameon=False)
plt.show()
```
