## Exercise 01.3


Per valutare il valore $\pi$ nell'esperimento di Buffon, limitiamo il piano in uno spazio con due linee, una nell'origine e l'altra a una distanza d.L'ago attraversa le linee se:
 
+ l'ascissa del suo estremo inferiore è < 0
+ l'ascissa del suo estremo superiore è > d

Per simulare il lanco dell'ago si generano : l'ascissa del centro dell'ago, e due punti generici nel piano per ricavare una pendenza senza usare $\pi$.
Ottenuta la pendenza m si controlla la posizione degli estremi $$x_e = x_{centro} \pm \sqrt(\frac{m^2 L^2}{1 + m^2})$$
con L la metà della lunghezza dell'asta.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("data.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('#throws')
plt.ylabel(r'$\pi$')
plt.title(r"Valori di $\pi$ in funzione dei lanci")
plt.grid(True)
plt.show()
```

#### Per poter estrarre un angolo casuale uniformemente (inclinazione dell'ago), sono stati estratti due punti sul piano la quale congiungente determina un coefficiente angolare. I dati riportati, per il teorema del limite centrale,  verificano che effetivamente la distribuzione di m sia uniforme

```python
from scipy.optimize import curve_fit

def f(x,a,b,c):  
    return a * b/((x-c)**2 + b**2)

fig, ax = plt.subplots(1, 1, figsize=(7, 5))

omega = np.loadtxt("testM.dat", usecols=(0), delimiter=' ', unpack='true')
bins = np.linspace(-10, 10, 50)
plt.hist(
    omega,
    bins=bins,
    density=True
);
plt.xlabel(r'$m$')
plt.ylabel(r'$\rho(m)$')
plt.title(r'Distribution of 1000 values of m, extracted with the two point method')

#fit
conteggi, bordi, _ = ax.hist(omega, bins=bins, density=True, color='green')
centri = (bordi[:-1] + bordi[1:]) / 2
parametri, _ = curve_fit(f, centri, conteggi)
y = f(bins, *parametri)
ax.plot(bins, y, 'r-', label=r"fit : $a\frac{b}{(x-c)^2 + b^2}$")
ax.text(2, 0.2,f'a = {parametri[0]}, \n b = {parametri[1]},\n c = {parametri[2]}')
ax.legend()
plt.show()
```
