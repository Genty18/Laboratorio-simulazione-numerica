## Uniform distribution

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#definizione gaussiana per il fit
def f(x,a,b,c):  
    return a * np.exp(-b*(x-c)**2)

#dati usati
x = np.linspace(1, 6, 10000)
bins = np.linspace(1, 6, 100)
dati = [1, 2, 10, 100]

#creo immagine
fig, ax = plt.subplots(1, 4, figsize=(17, 3))
fig2, ax2 = plt.subplots(1, 1, figsize=(8, 5))
fig.suptitle(r'$S_N$ from UNIFORM distribution with N = 1, 2, 10, 100', fontsize=19)
fig2.suptitle(r'Fitting the N = 100 data', fontsize=19)


for i in range(4):
    #carico dati e plot istogrammi
    unif = np.loadtxt(f"datiN{dati[i]}.dat", usecols=(0), delimiter=' ', unpack='true')
    ax[i].hist(
        unif,
        bins=bins,
        density=True,
        color='blue'
    );
    #intestazioni
    ax[i].set_xlabel(r'$S_N$',fontsize=13)
    ax[i].set_ylabel(r'$\rho(S_N)$',fontsize=13)

    # #fit
unif2 = np.loadtxt(f"datiN100.dat", usecols=(0), delimiter=' ', unpack='true')
ax2.hist(
    unif2,
    bins=bins,
    density=True,
    color='green'
)
ax2.set_ylabel(r'$\rho(S_N)$',fontsize=13)
ax2.set_xlabel(r'$S_N$',fontsize=13)
conteggi, bordi, _ = ax2.hist(unif2, bins=bins, density=True, color='green')
centri = (bordi[:-1] + bordi[1:]) / 2
parametri, _ = curve_fit(f, centri, conteggi)
y = f(x, *parametri)
ax2.plot(x, y, 'r-', label=r"fit : $ae^{-b(x-c)^2}$")
ax2.text(4.5, 1.3,f'a = {parametri[0]}, \n b = {parametri[1]},\n c = {parametri[2]}')
ax2.legend(loc=(0.75, 0.7), frameon=False)
```

## Distribuzione Cauchy-Lorentz

```python
#definizione lorentz per il fit
def f(x,a,b,c):  
    return a * b/((x-c)**2 + b**2)

#dati usati
x = np.linspace(-6, 6, 10000)
bins = np.linspace(-6, 6, 100)
dati = [1, 2, 10, 100]

#creo immagine
fig, ax = plt.subplots(1, 4, figsize=(17, 3))
fig2, ax2 = plt.subplots(1, 1, figsize=(8, 5))
fig.suptitle(r'$S_N$ from CAUCHY-LORENTZ distribution with N = 1, 2, 10, 100', fontsize=19)
fig2.suptitle(r'Fitting the N = 100 data', fontsize=19)


for i in range(4):
    #carico dati e plot istogrammi
    unif = np.loadtxt(f"datiN{dati[i]}.dat", usecols=(2), delimiter=' ', unpack='true')
    ax[i].hist(
        unif,
        bins=bins,
        density=True,
        color='blue'
    );
    #intestazioni
    ax[i].set_xlabel(r'$S_N$',fontsize=13)
    ax[i].set_ylabel(r'$\rho(S_N)$',fontsize=13)

    # #fit
unif2 = np.loadtxt(f"datiN100.dat", usecols=(2), delimiter=' ', unpack='true')
ax2.hist(
    unif2,
    bins=bins,
    density=True,
    color='green'
)
ax2.set_ylabel(r'$\rho(S_N)$',fontsize=13)
ax2.set_xlabel(r'$S_N$',fontsize=13)
conteggi, bordi, _ = ax2.hist(unif2, bins=bins, density=True, color='green')
centri = (bordi[:-1] + bordi[1:]) / 2
parametri, _ = curve_fit(f, centri, conteggi)
y = f(x, *parametri)
ax2.plot(x, y, 'r-', label=r"fit : $a\frac{b}{(x-c)^2 + b^2}$")
ax2.text(2, 0.2,f'a = {parametri[0]}, \n b = {parametri[1]},\n c = {parametri[2]}')
ax2.legend(loc=(0.75, 0.7), frameon=False)
```

As we expected, the C-L distribution remains a C-L and does not get the gaussian trend.
