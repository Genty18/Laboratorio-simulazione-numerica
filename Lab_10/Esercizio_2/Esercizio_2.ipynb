Parallelize with MPI libraries your Genetic Algorithm code in order to solve the TSP by performing a *Genetic Search* with (up to) 11 **parallel GA searches of the optimal path**:
each node should perform an independent GA search, the so-called *Continent*, **but** every $N_{migr}$ generations the *Continents* should exchange their best individuals randomly.

il numero $N_{migr}$ è stato scelto $N_{migr}$=50 su una simulazione di mille generazioni, per un totale di 20 scambi casuali tra "continenti".

### Percorsi trovati indipendentemente (a meno di 20 scambi casuali dei migliori individui) da 11 processi (continenti) per i capoluoghi di provinci italiani:

```python
import numpy as np
import matplotlib.pyplot as plt

# Creare una griglia di 6 righe per 2 colonne per avere 11 figure
fig, axes = plt.subplots(6, 2, figsize=(20, 48))
axes = axes.ravel()  # Flatten l'array 2D di assi in un array 1D per iterare facilmente

# Caricare le coordinate delle città una volta
x, y = np.loadtxt("OUTPUT/Cities.dat", usecols=(0, 1), unpack=True, skiprows=1)

# Iterare attraverso le 11 cartelle CONTINENT
for i in range(11):
    ax = axes[i]  # Selezionare l'asse corrente
    
    # Caricare la sequenza dalla cartella corrispondente
    sequence = np.loadtxt(f"OUTPUT/CONTINENT{i+1}/Sequences.dat", usecols=(0), unpack=True, skiprows=1)
    
    # Scatter plot delle città
    ax.scatter(x, y, label='Cities')
    
    # Creare le liste ordinate per il percorso migliore
    sorted_x = []
    sorted_y = []
    for j in sequence:
        sorted_x.append(x[int(j)-1])
        sorted_y.append(y[int(j)-1])
    sorted_x.append(x[int(sequence[0])-1])
    sorted_y.append(y[int(sequence[0])-1])
    
    # Tracciare il percorso migliore
    ax.plot(sorted_x, sorted_y, color='lightblue', linestyle='--', label='Best path')
    
    # Aggiungere titoli e etichette
    ax.set_title(f"Best GA's path of the TSP, for italian cities for the {i+1} process")
    ax.set_xlabel(r'$cities_x$')
    ax.set_ylabel(r'$cities_y$')
    
    # Mostrare una griglia
    ax.grid(True)
    
    # Aggiungere una legenda
    ax.legend()

# Nascondere eventuali assi vuoti (nel caso di 11 grafici)
for j in range(12):
    if j >= 11:
        fig.delaxes(axes[j])

# Aggiungere una spaziatura tra i subplot
plt.tight_layout()

# Visualizzare il grafico
plt.show()
```
