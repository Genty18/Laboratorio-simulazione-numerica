 ### Circonferenza

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Creare il grafico
fig, ax1 = plt.subplots(1, 1, figsize=(10, 8))

x, y = np.loadtxt("OUTPUT/CIRCUMFERENCE/Cities.dat", usecols=(0,1),unpack=True, skiprows=1)
sequence = np.loadtxt("OUTPUT/CIRCUMFERENCE/Sequences.dat", usecols=(0),unpack=True, skiprows=1)

plt.scatter(x, y, label='Cities')

sorted_x = []
sorted_y = []
for i in sequence:
    sorted_x.append(x[int(i)-1])
    sorted_y.append(y[int(i)-1])
sorted_x.append(x[0])
sorted_y.append(y[0])

# #RETURNING TO THE STARTING CITY
# sorted_x.append(x[int(sequence[0])])
# sorted_y.append(y[int(sequence[0])])
plt.plot(sorted_x, sorted_y, color='lightblue', linestyle='--', label='Best path')

# Aggiungere titoli e etichette
plt.title(r"Best GA's path of the TSP for 34 cities randomly placed on a CIRCUMFERENCE")
plt.xlabel(r'$cities_x$')
plt.ylabel(r'$cities_y$')

print("The shortest path's length has been found for the city reached as :\n ")
print(sequence)

# Mostrare una griglia
plt.grid(True)

# Aggiungere una legenda
plt.legend()

# Visualizzare il grafico
plt.show()
```

### $L^{(2)}$ per la circonferenza e il quadrato 

```python
import matplotlib.pyplot as plt
import numpy as np

# Creazione della figura con 2 subplots
fig, ax1 = plt.subplots(1, 2, figsize=(14, 6))

# Caricamento dei dati per il primo subplot (Square)
gen_square, l_square = np.loadtxt("OUTPUT/SQUARE/Generations.dat", usecols=(0, 1), unpack=True, skiprows=1)

# Plot del grafico per il primo subplot (Square)
ax1[0].plot(gen_square, l_square, label='Length')
ax1[0].set_xlabel('Generations')
ax1[0].set_ylabel(r'$L^{(2)}$')
ax1[0].set_title(r'$L^{(2)}$ values of Square as function of generations')
ax1[0].grid(True)
ax1[0].legend()

# Caricamento dei dati per il secondo subplot (Circumference)
gen_circ, l_circ = np.loadtxt("OUTPUT/CIRCUMFERENCE/Generations.dat", usecols=(0, 1), unpack=True, skiprows=1)

# Plot del grafico per il secondo subplot (Circumference)
ax1[1].plot(gen_circ, l_circ, label='Length')
ax1[1].set_xlabel('Generations')
ax1[1].set_ylabel(r'$L^{(2)}$')
ax1[1].set_title(r'$L^{(2)}$ values of Circumference as function of generations')
ax1[1].grid(True)
ax1[1].legend()

# Mostra una griglia solo nei subplots, non sulla figura intera
ax1[0].grid(True)
ax1[1].grid(True)


# Mostra il grafico completo
plt.tight_layout()
plt.show()
```
